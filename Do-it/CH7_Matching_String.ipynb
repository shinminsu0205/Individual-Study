```python
# 브루트 포스법으로 문자열 검색하기

def bf_match(txt: str, pat: str) -> int:
    pt = 0 #pointer-text
    pp = 0 #pointer-pattern

    while pp != len(txt) and pp != len(pat):
        if txt[pt] == pat[pp]:
            pt +=1
            pp +=1
        else:
            pt =pt - pp +1
            pp = 0
    
    return pt - pp if pp == len(pat) else -1

if __name__ == '__main__':
    s1 = input('텍스트를 입력하세요')
    s2 = input('패턴을 입력하세요')

    idx = bf_match(s1, s2)

    if idx == -1:
        print('텍스트 안에 패턴이 존재하지 않습니다')
    else:
        print(f'{idx+1}번째 문자가 일치합니다.')
```

```python
#KMP법으로 문자열 검색하기

def kmp_match(txt: str, pat: str)->int:
    pt = 1
    pp = 0
    skip = [0] * (len(pat)+1) #건너뛰기 표

    #건너뛰기 표 만들기
    skip[pt] = 0
    while pt != len(pat):
        if pat[pt] == pat[pp]:
            pt +=1
            pp +=1
            skip[pt] = pp
        elif pp == 0:
            pt+=1
            skip[pt] = pp
        else:
            pp = skip[pp]

    #문자열 검색하기
    pt = pp = 0
    while pt!=len(txt) and pp != len(pat):
        if txt[pt] == pat[pp]:
            pt +=1
            pp +=1
        elif pp == 0:
            pt +=1
        else:
            pp = skip[pp]
        
    return pt - pp if pp == len(pat) else -1

if __name__ == '__main__':
    s1 = input('텍스트를 입력하세요')
    s2 = input('패턴을 입력하세요')

    idx = kmp_match(s1, s2)

    if idx == -1:
        print('텍스트 안에 패턴이 존재하지 않습니다')
    else:
        print(f'{idx+1}번째 문자가 일치합니다.')
```

```python
#보이어, 무어법

def bm_match(txt: str, pat: str)->int:
    skip = [None]*256

    #건너뛰기 표 만들기
    for pt in range(256):
        skip[pt] = len(pat)
    for pt in range(len(pat)):
        skip[ord(pat[pt])] = len(pat) - pt - 1 #ord: 아스키 코드 받는 함수

    #검색하기
    while pt < len(txt):
        pp = len(pat) - 1
        while txt[pt] == pat[pp]:
            if pp == 0:
                return pt
            pt-=1
            pp-=1
        pt += skip[ord(txt[pt])] if skip[ord(txt[pt])] > len(pat) -pp else len(pat) - pp
    
    return -1

if __name__ == '__main__':
    s1 = input('텍스트를 입력하세요')
    s2 = input('패턴을 입력하세요')

    idx = bm_match(s1, s2)

    if idx == -1:
        print('텍스트 안에 패턴이 존재하지 않습니다')
    else:
        print(f'{idx+1}번째 문자가 일치합니다.')
```
